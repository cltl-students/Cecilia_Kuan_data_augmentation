# Data Statistics - qty per level in each category

```python
import pandas as pd
```

```python
gpt_pkl = '../data/gpt/gpt_data_all_updated.pkl'
df_gpt = pd.read_pickle(gpt_pkl)
df_gpt
```

```python
df_counts2 =  pd.DataFrame()
df_counts = pd.DataFrame()

# ADD TO df_counts2
#col = ['FAC', 'INS']
df_fac = df_gpt.loc[df_gpt['labels'] == 'FAC']
fac_qty = (df_fac['level'].value_counts().sort_index()).to_list()

df_ins = df_gpt.loc[df_gpt['labels'] == 'INS']
ins_qty = (df_ins['level'].value_counts().sort_index()).to_list()

df_counts2['INS'] = ins_qty
df_counts2['FAC'] = fac_qty

# ADD TO df_counts
#col2 = ['ADM', 'ATT', 'BER', 'ENR', 'ETN', 'MBW', 'STM']
df_adm = df_gpt.loc[df_gpt['labels'] == 'ADM']
adm_qty = (df_adm['level'].value_counts().sort_index()).to_list()

df_att = df_gpt.loc[df_gpt['labels'] == 'ATT']
att_qty = (df_att['level'].value_counts().sort_index()).to_list()

df_ber = df_gpt.loc[df_gpt['labels'] == 'BER']
ber_qty = (df_ber['level'].value_counts().sort_index()).to_list()

df_enr = df_gpt.loc[df_gpt['labels'] == 'ENR']
enr_qty = (df_enr['level'].value_counts().sort_index()).to_list()

df_ETN = df_gpt.loc[df_gpt['labels'] == 'ETN']
etn_qty = (df_ETN['level'].value_counts().sort_index()).to_list()

df_mbw = df_gpt.loc[df_gpt['labels'] == 'MBW']
mbw_qty = (df_mbw['level'].value_counts().sort_index()).to_list()

df_stm = df_gpt.loc[df_gpt['labels'] == 'STM']
stm_qty = (df_stm['level'].value_counts().sort_index()).to_list()

df_counts['ADM'] = adm_qty
df_counts['ATT'] = att_qty
df_counts['BER'] = ber_qty
df_counts['ENR'] = enr_qty
df_counts['MBW'] = mbw_qty
df_counts['STM'] = stm_qty
```

```python
final_qty = pd.concat([df_counts,df_counts2],axis=1)
final_qty = final_qty.sort_index(axis=1)
final_qty = final_qty.fillna(0)
final_qty['ADM'] = final_qty['ADM'].astype(int)
final_qty['ATT'] = final_qty['ATT'].astype(int)
final_qty['BER'] = final_qty['BER'].astype(int)
final_qty['ENR'] = final_qty['ENR'].astype(int)
final_qty['ENT'] = final_qty['ENT'].astype(int)
final_qty['MBW'] = final_qty['MBW'].astype(int)
final_qty['STM'] = final_qty['STM'].astype(int)
final_qty
```

```python
final_qty.to_csv('../data/gpt/gpt_qty_per_level.csv')
```
